```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/home/bj23811607/Downloads/flight_delays(Sheet1).csv")

pivot = df.pivot_table(index=['Destinations'], columns=['Airline', 'Status'], values='Flights')

delay_pct = df[df['Status'] == 'delayed'].groupby('Airline')['Flights'].sum() / df.groupby('Airline')['Flights'].sum()


pivot.plot(kind='bar', figsize=(12,6))
plt.title('Flight Arrival Stats by Airline and Destination')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Using a virtual Linux environment and Jupyter Notebook, I analyzed flight arrival data for two major airlines, ALASKA and AM WEST, across Los Angeles, Phoenix, San Diego, San Francisco, and Seattle. The chart revealed several notable patterns. AM WEST had a significantly higher volume of flights to Phoenix, with nearly 5,000 on-time arrivals—far surpassing any other destination in the dataset. This suggests Phoenix may serve as a major operational hub for AM WEST. ALASKA, while operating fewer flights, maintains a more balanced distribution and relatively lower delay rates.
