```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ["mpg", "cylinders", "displacement", "horsepower", "weight",
                "acceleration", "model_year", "origin", "car_name"]

df = pd.read_csv(url, delim_whitespace=True, names=column_names, na_values='?')

df["horsepower"] = pd.to_numeric(df["horsepower"], errors='coerce')

df["origin"] = df["origin"].map({1: "USA", 2: "Asia", 3: "Europe"})

sns.countplot(x="cylinders", data=df)
plt.title("Distribution of Cylinders")
plt.xlabel("Cylinders")
plt.ylabel("Count")
plt.show()

sns.scatterplot(x="weight", y="horsepower", hue="origin", data=df)
plt.title("Horsepower vs. Weight by Origin")
plt.xlabel("Weight")
plt.ylabel("Horsepower")
plt.show()

# Question: Does the relationship between horsepower and weight vary by region?

sns.lmplot(x="weight", y="horsepower", hue="origin", data=df, aspect=1.5)
plt.title("Horsepower vs. Weight with Regression Lines by Origin")
plt.xlabel("Weight")
plt.ylabel("Horsepower")
plt.show()
```

This week, I used Python to explore a car dataset called Auto MPG. I started by loading the data straight from the internet and cleaned it by fixing missing values in the horsepower column and changing the origin numbers into names like USA, Asia, and Europe. Then I made a bar chart to show how many cars had different cylinder counts. Most cars had 4 cylinders, while very few had 3 or 5. I also made scatter plots to look at the link between horsepower and weight. One of the questions I asked for myself was whether this link changes depending on where the car was made. The graphs showed that cars from the USA are usually heavier and have more horsepower, while cars from Europe and Asia are lighter and less powerful. These patterns helped me understand how car designs differ across regions.
