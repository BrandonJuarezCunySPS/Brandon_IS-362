```python
import pandas as pd
import numpy as np

df = pd.read_csv("/home/bj23811607/Downloads/movie_ratings(Sheet1).csv")

df.set_index('User', inplace=True)

print("Raw Ratings DataFrame:")
print(df)

user_avg = df.mean(axis=1)
movie_avg = df.mean(axis=0)

print("\n Average Ratings per User:")
print(user_avg)

print("\n Average Ratings per Movie:")
print(movie_avg)

def normalize(row):
    min_val = row.min()
    max_val = row.max()
    return (row - min_val) / (max_val - min_val)

normalized_df = df.apply(normalize, axis=1)

print("\n Normalized Ratings DataFrame:")
print(normalized_df)

norm_user_avg = normalized_df.mean(axis=1)
norm_movie_avg = normalized_df.mean(axis=0)

print("\n Normalized Average Ratings per User:")
print(norm_user_avg)

print("\n Normalized Average Ratings per Movie:")
print(norm_movie_avg)
```

Conclusion:
When we normalize movie ratings, we adjust each person’s scores so they fit on the same scale, usually from 0 to 1. This helps us compare how people feel about movies, even if they rate things differently. For example, someone who always gives high scores and someone who’s more critical can still be compared fairly. One big advantage of normalization is that it removes personal bias. It doesn’t matter if someone tends to rate everything a 5 or a 2 as we’re looking at how much they liked each movie compared to their own usual ratings. This is really helpful when building systems that recommend movies based on similar tastes.

But there are also downsides. When we normalize ratings, we lose the actual meaning of the numbers. A normalized score of 1.0 might mean a 5 star rating for one person, but only a 3-star rating for another. That can be confusing if we care about how much someone truly liked a movie, not just how it compares to their other ratings. So in short, normalization is great for finding patterns and comparing people’s preferences, but it’s not perfect if we want to understand their real opinions. It’s a tool, and like any tool, it works best when used in the right way. Even if I use a virtual Linux machine for Jupyter.

```python
def standardize(row):
    mean = row.mean()
    std = row.std()
    return (row - mean) / std

standardized_df = df.apply(standardize, axis=1)

print("\n Standardized Ratings DataFrame:")
print(standardized_df)

std_user_avg = standardized_df.mean(axis=1)
std_movie_avg = standardized_df.mean(axis=0)

print("\n Standardized Average Ratings per User:")
print(std_user_avg)

print("\n Standardized Average Ratings per Movie:")
print(std_movie_avg)


```
